```python
import pandas as pd
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import numpy as np
filename = 'C://train_r//data//data1_1//data1_1.csv'

data = pd.read_csv(filename)
```

```python
data = data.fillna(0)
data = data.replace('Y', 1)
data.label.replace(['기타질병', '비브리오병', '연쇄구균병', '스쿠티카병', 
       '정상', '활주세균병', '에드워드병', 'VHSV'], [0, 1, 2, 3, 4, 5, 6, 7], inplace=True)
```

```python
data
```

```python
from pycaret.classification import *
setup_clf = setup(data=data, target='label')
```

```python
rf = create_model('rf', fold=5)
```

```python
top5 = compare_models(sort='Accuracy', n_select=5)
```

```python
tuned_top5 = [tune_model(i) for i in top5]
```

```python
blender_top5 = blend_models(estimator_list=tuned_top5)
```

```python
final_model = finalize_model(blender_top5)
prediction = predict_model(final_model, data=data.iloc[-100:])
```

```python
from pycaret.utils import check_metric
check_metric(prediction['label'], prediction['Label'], metric = 'Accuracy')
```
