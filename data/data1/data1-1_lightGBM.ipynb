```python
import pandas as pd
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import numpy as np
filename = 'C://train_r//data//data1_1//data1_1.csv'

data = pd.read_csv(filename)
data
```

```python
data = data.fillna(0)
data = data.replace('Y', 1)
data.label.replace(['기타질병', '비브리오병', '연쇄구균병', '스쿠티카병', 
       '정상', '활주세균병', '에드워드병', 'VHSV'], [0, 1, 2, 3, 4, 5, 6, 7], inplace=True)
```

```python
import lightgbm as lgb
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 데이터 분리
train_data, test_data, train_labels, test_labels = train_test_split(data.drop("label", axis=1), data["label"], test_size=0.2)

# LightGBM 모델 파라미터 설정
params = {
    "objective": "binary",
    "boosting_type": "gbdt",
    "num_leaves": 31,
    "learning_rate": 0.05,
    "feature_fraction": 0.9,
    "bagging_fraction": 0.8,
    "bagging_freq": 5,
    "verbosity": -1
}

# LightGBM 모델 학습
lgbm = lgb.LGBMClassifier(**params)
lgbm.fit(train_data, train_labels)

# LightGBM 모델 평가
predictions = lgbm.predict(test_data)
accuracy = accuracy_score(test_labels, predictions)
print("Accuracy: {:.4f}".format(accuracy))
```
