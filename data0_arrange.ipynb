```python
import seaborn as sns
import pandas as pd

data_1 = data.groupby(by=['질병'], as_index=False).count()

total_rows = data.shape[0]
for i in ['VHSV', '기타질병', '비브리오병', '스쿠티카병', '에드워드병', '연쇄구균병', '정상', '활주세균병']:
    count = data_1.loc[data_1['질병'] == i, :].iloc[0, 1:]
    percent = (count / total_rows) * 100
    print(f"질병 = {i}:")
    print(percent)
    print("\n")
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

grouped = data.groupby(by=['질병'], as_index=False).sum()
total_rows = data.shape[0]

for i in ['VHSV', '기타질병', '비브리오병', '스쿠티카병', '에드워드병', '연쇄구균병', '정상', '활주세균병']:
    count = grouped.loc[grouped['질병'] == i, :].iloc[0, 1:]
    percent = (count.sum() / total_rows) * 100
    print(f"질병 = {i}:")
    print(percent)
    print("\n")
```
